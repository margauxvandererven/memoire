```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks

# Spectres synthétique et observé (exemples fictifs)
# wavelength_synth, flux_synth : longueur d'onde et flux du spectre synthétique
# wavelength_obs, flux_obs : longueur d'onde et flux du spectre observé
wavelength_synth = np.array(syntspec(path+filename2)['wavelen'][1500:100000])  # Longueurs d'onde du spectre synthétique
flux_synth = np.array(diff[:-25500])        # Flux du spectre synthétique
wavelength_obs = np.array(redshift_wavelen(stardata.get("wavelen_h"), stardata.get("v_h"))[:40000])    # Longueurs d'onde du spectre observé
flux_obs = np.array(stardata.get("flux_h")[:40000])          # Flux du spectre observé

# Inversion du flux pour détecter les pics d'absorption
# On détecte les pics sur -flux pour les absorptions
peaks_synth, _ = find_peaks(-flux_synth, prominence=0.1)  # Ajuster `prominence` selon le niveau de bruit
peaks_obs, _ = find_peaks(-flux_obs, prominence=0.1)

# Longueurs d'onde des pics d'absorption
wavelength_peaks_synth = wavelength_synth[peaks_synth]
flux_peaks_synth = flux_synth[peaks_synth]
wavelength_peaks_obs = wavelength_obs[peaks_obs]
flux_peaks_obs = flux_obs[peaks_obs]

# Tolérance de correspondance en angströms
tolerance_wavelength = 1  # Tolérance en angströms
tolerance_flux = 0.05         # Tolérance en flux (à ajuster selon l'intensité)

# Recherche de correspondances
matching_peaks = []
for ws, fs in zip(wavelength_peaks_synth, flux_peaks_synth):
    # Recherche des pics dans le spectre observé proches en longueur d'onde ET en flux
    matches = [
        (wo, fo) for wo, fo in zip(wavelength_peaks_obs, flux_peaks_obs)
        if abs(wo - ws) <= tolerance_wavelength and abs(fo - fs) <= tolerance_flux
    ]
    if matches:
        matching_peaks.append((ws, fs, matches))

# Affichage des correspondances trouvées
print("Correspondances trouvées (synthétique, observé) :")
for match in matching_peaks:
    print("Longueur d'onde synthétique : {:.6f} Å, Flux synthétique : {:.6f}, Correspondance(s) observée(s) : {}".format(
        match[0], match[1], [(m[0], m[1]) for m in match[2]]
    ))
# Visualisation
plt.figure(figsize=(25, 6))
plt.plot(wavelength_synth, flux_synth, label="Spectre Synthétique", color="blue", alpha=0.7, linewidth=0.2)
plt.plot(wavelength_obs, flux_obs, label="Spectre Observé", color="orange", alpha=0.7, linewidth=0.2)

# Mettre en évidence les pics
plt.scatter(wavelength_peaks_synth, flux_synth[peaks_synth], color="blue", marker='x', label="Pics Synthétiques")
plt.scatter(wavelength_peaks_obs, flux_obs[peaks_obs], color="orange", marker='o', label="Pics Observés", linewidths=0.01)

# Mettre en évidence les pics correspondants
for match in matching_peaks:
    for wo, fo in match[2]:
        # Lignes pointillées entre les pics correspondants
        plt.plot([match[0], wo], [match[1], fo], 'g--')

plt.xlabel("Longueur d'onde (Å)")
plt.ylabel("Flux")
plt.legend()
plt.title("Détection des pics d'absorption et correspondances entre spectres")
plt.show()
```
